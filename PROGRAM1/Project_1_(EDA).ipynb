**Step 1: Import Libraries**

```python
# Dataset: Titanic
# Skills Used: Python, Pandas, Matplotlib

# import libraries
import pandas as pd
import matplotlib.pyplot as plt

# pandas is used for data handling
# matplotlib is used for graphs (visualization)
```

**Step 2: Load dataset**

```python
# load dataset from online source

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

# show first 5 rows
df.head()
```

**Step 3: Understand the data**

```python
# number of rows and columns
df.shape

# column names and data types
df.info()

# statistical summary
df.describe()
```

**Step 4: Check Missing Values**

```python
# count missing values
df.isnull().sum()
```

```python
#Handle Missing Values

# fill missing Age iwth average age
df["Age"].fillna(df["Age"].mean(), inplace=True)

# fill missing Embarked with most common Value
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

**Step 5: Simple Analysis**

```python
# Survival count

df["Survived"].value_counts()
```

**Step 6: Data Visualization**

```python
# Survival Bar Chart

df["Survived"].value_counts().plot(kind="bar")
plt.title("Survival Count")
plt.xlabel("Survived (0= No, 1=Yes)")
plt.ylabel("Number of passenger")

plt.show()
```

```python
# Gender vs Survival

df.groupby("Sex")["Survived"].mean().plot(kind="bar")
plt.title("Survival Count")
plt.ylabel("Survived (0= No, 1= Yes)")
plt.show()
```

```python
# Age Distribution

plt.hist(df["Age"], bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

We've seen this in the project 1:
*   More passengers died than survived
*   Females had a higher survival rate
*   Most passengers were between 20-40 years old
*   Age and gender strongly affected survival
